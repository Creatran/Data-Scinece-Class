## b. Introduction and background
### About the dataset
In this report, I concentrated on Wisconsin Diagnostic Breast Cancer (WDBC) dataset. Features in this dataset are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe the characteristics of the cell nuclei present in the image with 32 attributes. The dataset is available on the UCI machine learning repository (https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29).

Attributes:

1: Id number

2: Diagnosis: malignant; benign

3 - 32: ten real-valued features computed for each cell nucleus.

    a: radius

    b: texture (standard deviation of gray-scale values)

    c: perimeter

    d: area

    e: smoothness
    
    f: compactness
    
    g: concavity

    h: concave points

    i: symmetry

    j: fractal dimension

The mean, standard error, and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. For instance, field 3 is Mean Radius, field
13 is Radius SE, field 23 is the Worst Radius.
### About the external data set

1: Investigate which features have a correlation with each other and find out whether drop some features may have a huge influence on the classification. 

First, take a look at the correlation heatmap.

```python
# plt.figure(figsize=(18, 18))
corr = wdbc_features.corr()
sns.heatmap(corr)
plt.title("Heapmap")
plt.show()
```

The heatmap shows that there are some features strongly correlated with each other, so it is not necessary to use them all since one feature may present the same message with the other. For example: figure radius_mean, perimeter_mean and area_mean are correlated with each other. Compactness_mean, concavity_mean and concave points_mean are correlated with each other. Radius_se, perimeter_se and area_se are correlated. Radius_worst, perimeter_worst and area_worst are correlated. Compactness_worst, concavity_worst and concave points_worst are correlated. Compactness_se, concavity_se and concave points_se are correlated. Texture_mean and texture_worst are correlated. Area_worst and area_mean are correlated.


I kept one feature and dropped the others in each group where features with each other have a correlation higher than 0.9, and saved the data into wdbc_Selected.

```python
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
selected_columns = wdbc_features.columns[columns]
wdbc_Selected = wdbc_features[selected_columns]
```

Overall, we have 20 selected features that are not have a correlation higher than 0.9. Plot all the selected features.

```python
fig = plt.figure(figsize = (20, 25))
j = 0
for i in wdbc_Selected.columns:
    plt.subplot(6, 4, j+1)
    j += 1
    sns.distplot(wdbc_Selected[i][wdbc['diagnosis']=='B'], color='g', label = 'benign')
    sns.distplot(wdbc_Selected[i][wdbc['diagnosis']=='M'], color='r', label = 'malignant')
    plt.legend(loc='best')
fig.suptitle('Breast Cance Data Analysis')
fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
```

The plots show the 20 selected features that do not have any correlation with each other. But some of them might have little contribution in classification. As shown in the plot, the texture_se, smoothness_se, symmetry_se, fractal dimension_se features do not attribute much to the classification of benign or malignant.

Next, I will do PCA on both the original dataset and the selected dataset and compare their outcomes.

```python
wdbc_y = wdbc_diag.replace(['M', 'B'], [1, 0])

from sklearn.decomposition import PCA
import timeit

start = timeit.default_timer()

pca1 = PCA()
pca1.fit(wdbc_features)
print("Cumsum variance for original features:\n", np.cumsum(pca1.explained_variance_ratio_))

stop = timeit.default_timer()
print('Time: ', stop - start)  
```

```python
start = timeit.default_timer()

pca2 = PCA()
pca2.fit(wdbc_Selected)
print("Cumsum variance for selected features:\n", np.cumsum(pca2.explained_variance_ratio_))

stop = timeit.default_timer()
print('Time: ', stop - start)  
```

Compare the outcomes and we can see that dropping the correlation features do not lower much of PCA's efficacy, but it does minimize the computation time.

```python
plt.plot(np.arange(len(pca2.explained_variance_ratio_)), np.cumsum(pca2.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title('Cumulative Explained Variance plot')
plt.show()
```

2: Clustering all the patients into two groups by their features and find out how well is the classitication towards tumor type.

First, use k-means to cluster all the features and see the classification correct rate.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2)
kmeans.fit(wdbc_norm)
y_kmeans = kmeans.predict(wdbc_norm)

print("classification correct rate: ", 1 - sum(abs(y_kmeans  - wdbc_y['diagnosis']))/len(y_kmeans))
```

The classification works quite well, with a correct rate of 91.04%. Besides, I want to see what would happen if I cluster on the three groups.

```python
kmeans_mean = KMeans(n_clusters = 2)
kmeans_mean.fit(wdbc_norm[features_mean])
y_kmeans_mean = kmeans_mean.predict(wdbc_norm[features_mean])
print("Group 1 classification correct rate: ", 1 - sum(abs(y_kmeans_mean  - wdbc_y['diagnosis']))/len(y_kmeans_mean))
```

```python
kmeans_se = KMeans(n_clusters = 2)
kmeans_se.fit(wdbc_norm[features_se])
y_kmeans_se = kmeans_se.predict(wdbc_norm[features_se])
print("Group 2 classification correct rate: ", 1 - sum(abs(y_kmeans_se  - wdbc_y['diagnosis']))/len(y_kmeans_se))
```

```python
kmeans_worst = KMeans(n_clusters = 2)
kmeans_worst.fit(wdbc_norm[features_worst])
y_kmeans_worst = kmeans_worst.predict(wdbc_norm[features_worst])
print("Group 3 classification correct rate: ", 1 - sum(abs(y_kmeans_worst  - wdbc_y['diagnosis']))/len(y_kmeans_worst))
```

The above classification results show that group1(mean group) and group3(worst group) have a better classification rate than group2. So, the diagnosis outcome has a stronger relationship with mean value and worst value of tumors.

3: Merge the WDBC and WPBC dataset and find out if there is a relationship between the recurrence time and each features.

Firstly, do some modefication of the outside data and merge the two dataset. Save the merged data in merged_data.

```python
wpbc_outcome = wpbc.iloc[:, 0:3]
merged_data = pd.merge(wpbc_outcome, wdbc, on = 'id')
merged_data['outcome'] = merged_data['outcome'].replace(['R', 'N'], [0, 1])
merged_data['diagnosis'] = merged_data['diagnosis'].replace(['M', 'B'], [1, 0])
```

Group the data by outcome, and find that there are more patients with recurrence(104) than without recurrence(35).

```python
merged_data.groupby('outcome')['outcome'].count()
```

Draw the distribution plot for recurrence time and Non-recurrence time. From the plot we could see that the the mean value for non-recurrence time is greater than that of recurrence time. 

```python
rec = merged_data['time'][merged_data.outcome == 1]
notrec = merged_data['time'][merged_data.outcome == 0]

sns.distplot(rec, color='g', label = 'Recurrence time')
sns.distplot(notrec, color='r', label = 'Not recurrence time')
plt.title("Time")
plt.ylabel("Frequency")
plt.legend(loc='best')
plt.show()
```

Devide the dataset into two group by outcome. Applied two sample t-test on each features with the two group and selected the features with a p-value less than 0.05.

```python
group1 = merged_data[merged_data.outcome == 0][merged_data.columns[2:]]
group2 = merged_data[merged_data.outcome == 1][merged_data.columns[2:]]

from scipy.stats import ttest_ind, ttest_ind_from_stats
from scipy.special import stdtr

t = pd.DataFrame()
for i in merged_data.columns[2:]:
    a, b = ttest_ind(group1[i], group2[i])
    if b < 0.05:
        t[i] = a, b, group1[i].mean() - group2[i].mean()
        
t = t.rename(index = {0: 'test_statistics', 1: 'p_value', 2: 'Diff in mean'})
t
```

Shown from the result table, time and other 7 features showed statistically significant differences between the two groups. So, for patients with larger mean and worst value of radius, perimeter and area, and has lower mean fractal dimension value tend to have larger possibility to have tumor recurrence.
